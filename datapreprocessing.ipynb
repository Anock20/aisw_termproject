```python
import json 
from collections import Counter
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows 환경: 맑은 고딕
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# JSON 데이터 로드 (파일 경로에 맞게 변경)
with open(r"E:\Desktop\lee\3g\aisoftware\term\문서요약 텍스트\Training\신문기사_train_original\train_original.json", 'r', encoding='utf-8') as file:
    data = json.load(file)

# documents 내부의 데이터에서 category 추출
documents = data.get("documents", [])  # "documents" 키에서 데이터를 가져옴
categories = [doc.get("category") for doc in documents]

# 각 category의 개수 계산   
category_counts = Counter(categories)

# 리스트로 변환
category_list = [{"category": cat, "count": count} for cat, count in category_counts.items()]
print(category_list)  # 리스트 출력

# 데이터 시각화
categories = list(category_counts.keys())
counts = list(category_counts.values())

plt.figure(figsize=(10, 6))
plt.bar(categories, counts, color='skyblue')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Category Counts', fontsize=15)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import json

# JSON 데이터 로드 (파일 경로에 맞게 변경)
with open(r"E:\Desktop\lee\3g\aisoftware\term\문서요약 텍스트\Training\신문기사_train_original\train_original.json", 'r', encoding='utf-8') as file:
    data = json.load(file)

# "documents" 배열에서 "category": "경제"인 항목만 추출
documents = data.get("documents", [])  # documents 배열 가져오기
economy_data = [item for item in documents if item.get("category") == "경제"]

# 결과 확인
print(json.dumps(economy_data, ensure_ascii=False, indent=4))

# 선택적으로 결과를 새 파일에 저장
with open('economy_data.json', 'w', encoding='utf-8') as outfile:
    json.dump(economy_data, outfile, ensure_ascii=False, indent=4)
```

```python
import json

# JSON 데이터 로드
with open("economy_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 데이터 가공
processed_data = []
for article in data:
    # 제목과 본문을 연결하여 입력 텍스트 생성
    title = article["title"]
    text_sentences = []
    for paragraph in article["text"]:
        for sentence_data in paragraph:
            text_sentences.append(sentence_data["sentence"])
    full_text = " ".join(text_sentences)  # 본문 문장들을 공백으로 연결

    # 요약 텍스트 추출
    abstractive_summary = article["abstractive"][0]  # 첫 번째 요약만 사용

    # 처리된 데이터 저장
    processed_data.append({
        "input": f"제목: {title} 본문: {full_text}",
        "output": abstractive_summary
    })

# 가공된 데이터를 JSONL로 저장
with open("processed_dataset.jsonl", "w", encoding="utf-8") as f:
    for entry in processed_data:
        json.dump(entry, f, ensure_ascii=False)
        f.write("\n")
```
